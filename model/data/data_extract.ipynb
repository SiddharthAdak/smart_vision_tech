```python
def data_extracter(link,category,sub):
    
    service = Service(r"C:\Users\Atharva Sankhe\Downloads\chromedriver-win64\chromedriver-win64\chromedriver.exe")

    # Create a new instance of the Chrome driver using the Service object
    driver = webdriver.Chrome(service=service)
    wait = WebDriverWait(driver, 10)

    # Open a website
    driver.get(link)


    #putting pincode part 
    input_field = driver.find_element(By.CLASS_NAME, '_1UOrqe')
    input_field.send_keys("401305")
    input_field.submit()
    time.sleep(7)



    # getting products 

    # all rows
    product_rows = driver.find_elements(By.CLASS_NAME,'_75nlfW')

    #each row
    for row in product_rows:
        product_elements = row.find_elements(By.XPATH, './/div[@data-id]') 

        #each ele in row
        for product in product_elements:

            #get product name
            product_name_element = product.find_element(By.XPATH, './/a[@title]')  # Get the anchor tag with title
            product_name = product_name_element.get_attribute('title')
            print(product_name)

            #get product image link
            product_image_element = product.find_element(By.XPATH, './/img[@class="DByuf4"]')  # Get the image element
            product_image_link = product_image_element.get_attribute('src')
            print(product_image_link)

            #get discounted product price
            discounted_price_element = product.find_element(By.XPATH, './/div[@class="Nx9bqj GvWNMG"]')  # Discounted price
            discounted_price = discounted_price_element.text

            #get actual price
            actual_price_element = product.find_element(By.XPATH, './/div[@class="yRaY8j"]')  # Actual price
            actual_price = actual_price_element.text

            # Extract the discount percentage (if available)
            try:
                discount_percentage_element = product.find_element(By.XPATH, './/div[@class="UkUFwK jFaZKp"]/span')  # Discount percentage
                discount_percentage = discount_percentage_element.text
            except:
                discount_percentage = "N/A"  


            # append to list
            products.append({
                    'category':category,
                    'sub': sub,
                    'name': product_name,
                    'image_link': product_image_link,
                    'discounted_price': discounted_price,
                    'actual_price': actual_price,
                    'discount_percentage': discount_percentage
                })
        

    time.sleep(8)
    driver.quit()
    
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# Initialize the WebDriver
driver = webdriver.Chrome(executable_path='path_to_chromedriver')

# Open the product page
driver.get(r'https://www.flipkart.com/grocery/snacks-beverages/biscuits/pr?sid=73z,ujs,eb9&otracker=categorytree&marketplace=GROCERY&fm=neo%2Fmerchandising&iid=M_453ba81c-f401-456d-aa63-8f5eaa143999_2_G211C67CJ4GB_MC.FDQNSSB6Y0LQ&otracker=dynamic_rich_navigation_1_2.navigationCard.RICH_NAVIGATION_Snacks%2B%2526%2BBeverages~Biscuits~All_FDQNSSB6Y0LQ&otracker1=dynamic_rich_navigation_PINNED_neo%2Fmerchandising_NA_NAV_EXPANDABLE_navigationCard_cc_1_L2_view-all&cid=FDQNSSB6Y0LQ')

# Wait for the product elements to load
wait = WebDriverWait(driver, 10)

try:
    # Wait until product rows are present
    product_rows = wait.until(EC.presence_of_all_elements_located((By.CLASS_NAME, 'cPHDOP.col-12-12')))

    # Initialize a list to store product data
    products = []

    # Loop through each product row
    for row in product_rows:
        # Extract the product name
        product_name_element = row.find_element(By.XPATH, './/a[@title]')  # Get the anchor tag with title
        product_name = product_name_element.get_attribute('title')  # Get the title attribute

        # Extract the product image link
        product_image_element = row.find_element(By.XPATH, './/img[@class="DByuf4"]')  # Get the image element
        product_image_link = product_image_element.get_attribute('src')  # Get the src attribute

        # Append the extracted data to the products list
        products.append({
            'name': product_name,
            'image_link': product_image_link
        })

    # Print the extracted product data
    for product in products:
        print(f"Product Name: {product['name']}, Image Link: {product['image_link']}")

except Exception as e:
    print("Error:", e)

finally:
    # Close the browser when done
    driver.quit()
```
